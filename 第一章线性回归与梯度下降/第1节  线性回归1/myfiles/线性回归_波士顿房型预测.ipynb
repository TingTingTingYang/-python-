```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family']='SimHei'
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.size']=12
```

```python
boston=load_boston()
```

```python
boston.keys()
```

```python
boston.target
```

```python
X,y=boston.data,boston.target
df=pd.DataFrame(np.concatenate([X,y.reshape(-1,1)],axis=1),
               columns=boston.feature_names.tolist()+['MEDV'])
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['RM'],df['MEDV'],test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
numerator=np.sum((x_train-x_train.mean())*(y_train-y_train.mean()))
```

```python
numerator
```

```python
w1=9.31
w0=-36.18
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(6,4))
plt.scatter(x_train,y_train,label='训练集',c='g',marker='.',s=10)
plt.scatter(x_test,y_test,label='测试集',c='orange',marker='*',s=10)
x=np.linspace(x_train.min()-0.5,x_train.max()+0.5,100)
y=w0+w1*x
plt.plot(x,y,c='r',label='回归方程')
plt.xlabel('平均房间数')
plt.ylabel('房屋价格')
plt.legend()
plt.title(f'回归方程：$y={w1:.2f}x+{w0:.2f}$')
#两侧的$加不加都可以的，加上是公式的格式。会更美观一些
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
x_train=x_train.values.reshape(-1,1)
x_test=x_test.values.reshape(-1,1)
```

```python
lr.fit(x_train,y_train)
y_hat=lr.predict(x_test)
```
